# 1. Scrape the details of most viewed videos on YouTube from Wikipedia:

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 You need to find following details:
 A) Rank
 B) Name
 C) Artist
 D) Upload date
 E) Views

```python
#first connect to web driver
driver = webdriver.Chrome(r"C:\Users\WIN-10\Downloads\chromedriver.exe")
```

```python
#Getting Link
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
try:
    ranking_of_vid=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranking_of_vid:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('No details available')
except StaleElementReferenceException:
    Rank.append('No details available')
```

```python
Name=[]
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('No details available')
except StaleElementReferenceException:
    Name.append('No details available')
```

```python
Artist
try:
    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('No details available')
except StaleElementReferenceException:
    Artist.append('No details available')
```

```python
try:
    date_list=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date_list:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('No details available')
except StaleElementReferenceException:
    Upload_Date.append('No details available')
```

```python
try:
    list1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in list1:
        Views.append(i.text)
except NoSuchElementException:#handling no such element exception
    Views.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    Views.append('No details available')
```

```python
##creating and printing dataframe
df=pd.DataFrame({})
df['Rank']=Rank
df['Name']=Name
df['Artist']=Artist
df['Upload_date']=Upload_Date
df['Views']=Views[:30]
df
```

# 7. Scrape the details of Data science recruiters from naukri.com.

Url = https://www.naukri.com/ You have to find the following details: A) Name B) Designation C) Company D) Skills they hire for E) Location Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
#importing driver
##opening chrome window
driver = webdriver.Chrome(r"C:\Users\WIN-10\Downloads\chromedriver.exe")
#Getting Link
driver.get('https://www.naukri.com/')
```

```python
recr = driver.find_element_by_xpath("//ul[@class='midSec menu']/li[2]/a").get_attribute('href')
driver.get(recr)
```

```python
#finding job bt different element
search_job = driver.find_element_by_xpath("//div[@class='inpWrap']/input")
search_job.send_keys('Data Science')
```

```python
#clicking Search button
search_btn = driver.find_element_by_id("qsbFormBtn")
search_btn.click()
```

```python
names = []
des = []
loc = []
comp =[]
skills = []
```

```python
skill_tag = driver.find_elements_by_xpath("//div[@class='recInfo']/div[2]")
block = driver.find_elements_by_xpath("//p[@class='highlightable']")
```

```python
for i in block:
#Scrapping Name
    if len(i.text.split('\n')) == 4:
        try:
            names.append(i.text.split('\n')[0])
        except NoSuchElementException as e:
            names.append("--")
    else:
        names.append("--")
```

```python
#Scrapping Designation
for i in block:    
    if len(i.text.split('\n')) == 4:
        try:
            des.append(i.text.split('\n')[1])
        except NoSuchElementException as e:
            des.append("--")
    else:
        des.append("--")
```

```python
#Scrapping Company
for i in block:
    if len(i.text.split('\n')) == 4:
        try:
            comp.append(i.text.split('\n')[2])
        except NoSuchElementException as e:
            comp.append("--")
    else:
        comp.append("--")
```

```python
#Scrapping location
for i in block:
    if len(i.text.split('\n')) == 4:
        try:
            loc.append(i.text.split('\n')[3])
        except NoSuchElementException as e:
            loc.append("--")
    else:
        loc.append("--")
```

```python
#Scrapping Skills they hire for
for i in skill_tag:
    try:
        skills.append(i.text)
    except NoSuchElementException as e:
        skills.append("--")
```

```python
#making dataframe
df = pd.DataFrame({'Name':names,'Designation':des,'Company':comp,'Skills they hire for':skills,'location':loc})
```

```python
df
```

```python
driver.close()
```

```python
#importing driver
##opening chrome window
driver = webdriver.Chrome(r"C:\Users\WIN-10\Downloads\chromedriver.exe")
#Getting Link
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/')
```

```python
# Creating lists to store data

book=[]
author=[]
sold=[]
publisher=[]
genre=[]
```

```python
# Scraping data

for i in range(0,100):
    
# Scraping the book name
    
    b = driver.find_elements_by_id("table-cell-10943-{}-1".format(i))
    for j in b:
            book.append(j.text)
```

```python
# Scraping the author name
for i in range(0,100):
    a = driver.find_elements_by_id("table-cell-10943-{}-2".format(i))
    for j in a:
        author.append(j.text)
```

```python
 # Scraping the volumes sold
for i in range(0,100):
    s = driver.find_elements_by_id("table-cell-10943-{}-3".format(i))
    for j in s:
        sold.append(j.text)
```

```python
 # Scraping the publisher
for i in range(0,100):
    p = driver.find_elements_by_id("table-cell-10943-{}-4".format(i))
    for j in p:
        publisher.append(j.text)
```

```python
# Scraping the genre
for i in range(0,100):
    g = driver.find_elements_by_id("table-cell-10943-{}-5".format(i))
    for j in g:
        genre.append(j.text)
```

```python
# Creating dataframe

x= {'Book Name':book, 'Author Name': author, 'Volumes Sold': sold, 'Publisher':publisher, 'Genre': genre}
df= pd.DataFrame(x)
df
```
