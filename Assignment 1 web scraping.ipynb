# 9. Write a python program to scrape weather details for last 24 hours from ‘https://en.tutiempo.net/delhi.html?data=last-24- hours’ :

```python
#importing neccesary liberary
import  requests
import pandas as pd

#fetching data using BeautifulSoup
from  bs4 import BeautifulSoup
url = requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')
url
```

```python
# Weather condition
soup = BeautifulSoup(url.text, 'html.parser')
Weather_condition=[]
for i in soup.find_all ('td',class_='t Temp')[0:24]: 

    if i.previous_sibling is not None:

        Weather_condition.append(i.previous_sibling.text)

    else:

        Weather_condition.append(' ')
```

```python
# WIND
wind=[]
for i in soup.find_all ('td',class_='hr')[0:24]: 

    if i.previous_sibling is not None:

        wind.append(i.previous_sibling.text)

    else:

        wind.append(' ')
```

```python
# TEMPERATURE
Temp=[]
for i in soup.find_all ('td',class_='wind')[0:24]: 

    if i.previous_sibling is not None:

        Temp.append(i.previous_sibling.text)

    else:

        Temp.append(' ')
```

```python
# humdity
hum=[]
for i in soup.find_all ('td',class_='prob')[0:24]: 

    if i.previous_sibling is not None:

        hum.append(i.previous_sibling.text)

    else:

        hum.append(' ')
```

```python
#PRESSURE
pre=[]
for i in soup.find_all ('td',class_='prob')[0:24]:
    pre.append(i.text)
```

```python
# HOUR
Hour=[]
for i in soup.find_all ('td',class_='t Temp')[0:24]: 

    if i.previous_sibling.previous_sibling is not None:

        Hour.append(i.previous_sibling.previous_sibling.text)

    else:

        Hour.append(' ')
```

```python
#importing all the above data in DataFrame
df=pd.DataFrame({'Hour':Hour,'Temp':Temp,'Wind':wind,'Pressure':pre,'Humidity':hum})
df
```
